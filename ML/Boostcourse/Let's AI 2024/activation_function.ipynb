```python
# -*- coding: utf-8 -*-

import numpy as np
import matplotlib.pyplot as plt
import os

# 그래프 이미지 저장할 디렉토리명
SAVE_PNG_DNAME = 'images'
if os.path.exists(os.path.join(os.getcwd(), SAVE_PNG_DNAME)) == False:
    os.makedirs(os.path.join(os.getcwd(), SAVE_PNG_DNAME), mode=755, exist_ok=True)
os.chdir(os.path.join(os.getcwd(), f'{SAVE_PNG_DNAME}'))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha: float):
    return np.maximum(x, alpha*x)

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
%matplotlib inline
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

# 그래프 그리기
plt.plot(x, y)
plt.ylim(-0.1, 1.1)  # y축 범위 지정
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sigmoid Function')
plt.grid(True)
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = tanh(x)

plt.figure(facecolor='white')
plt.plot(x, y)
plt.plot([0,0],[1.0,-1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Tanh Function')
plt.tight_layout()	
plt.savefig("tanh_output.png")
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.figure(facecolor='white')
plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Relu Function')
plt.savefig('Relu_output.png')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x, alpha=0.05)

plt.figure(facecolor='white')
plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Leaky Relu Function')
plt.savefig('LeakyRelu_output.png')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) / np.sum(np.exp(x))

plt.figure(facecolor='white')
plt.plot(x, y)
plt.title('Softmax Function')
plt.savefig('softmax_output.png')
plt.show()
```
